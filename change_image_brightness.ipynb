```python
import pandas as pd
from PIL import Image, ImageEnhance
```

```python
data = pd.read_csv('changed_local.csv')
mnet = data[data['Mobilenet_v1'] == 1]
failed = data[((data['Mobilenet_v1'] == 0) & 
            (data['Inception_v4'] == 0) & 
            (data['Resnet_v1_152'] == 0))]
m_small = mnet.head(500)
m_large = mnet.head(5000)
f_small = failed.head(500)
f_large = failed.head(500)
```

# Test

```python
balanced_files = mnet['filename'].tolist()[:6500] + failed['filename'].tolist()[:6500]

im = Image.open("images/val/images/" + balanced_files[0])
enhancer = ImageEnhance.Brightness(im)

threehalf = enhancer.enhance(0.75)
half = enhancer.enhance(0.5)
quarter = enhancer.enhance(0.25)
tenth = enhancer.enhance(0.1)
im.show()
threehalf.show()
half.show()
quarter.show()
tenth.show()
```

# Decrease ImageNet Brightness

```python
files = data['filename'].tolist()
```

```python
for i in files:
    #read the image
    im = Image.open("images/val/images/" + i)

    #image brightness enhancer
    enhancer = ImageEnhance.Brightness(im)

    factor = 1 #gives original image
    im_output = enhancer.enhance(factor)
    im_output.save('balanced_images/original/'+i)

    factor = 0.75 #slight decreasse
    im_output = enhancer.enhance(factor)
    im_output.save('balanced_images/threehalf/'+i)
    
    factor = 0.5 #decreased by half
    im_output = enhancer.enhance(factor)
    im_output.save('balanced_images/half/'+i)

    factor = 0.25 #decreased by 75%
    im_output = enhancer.enhance(factor)
    im_output.save('balanced_images/quarter/'+i)
    
    factor = 0.1 #decreased by 90%
    im_output = enhancer.enhance(factor)
    im_output.save('balanced_images/tenth/'+i)
```
